```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#import dataset
df_train=pd.read_csv('train.csv')
df_train.head()
```

```python
#import test data
df_test=pd.read_csv('test.csv')
df_test.head()
```

```python
#Merge both train and test data
df=pd.concat([df_train,df_test])
```

```python
df.head()
```

```python
df.describe()
```

```python
df.drop(['User_ID'],axis=1,inplace=True)
```

```python
df.head()
```

```python
#handling Categorical feature gender
df['Gender']=df['Gender'].map({'F':0,'M':1})
df.head()
```

```python
#handling Categorical feature age
df['Age'].unique()
```

```python
df['Age']=df['Age'].map({'0-17':1,'18-25':2,'26-35':3,'36-45':4,'46-50':5,'51-55':6,'55+':7})
df.head()
```

```python
df_city=pd.get_dummies(df['City_Category'],drop_first=True)
df_city.head()
```

```python
df=pd.concat([df,df_city],axis=1)
df.head()
```

```python
#drop city category
df.drop('City_Category',axis=1,inplace=True)
df.head()
```

```python
## Mising values
df.isnull().sum()
```

```python
## Focus on replacing null values
df['Product_Category_2'].value_counts()
```

```python
##replace null values
df['Product_Category_2']=df['Product_Category_2'].fillna(df['Product_Category_2'].mode()[0])
```

```python
df['Product_Category_2'].isnull().sum()
```

```python
df['Product_Category_3'].unique()
```

```python
df['Product_Category_3'].value_counts()
```

```python
df['Product_Category_3']=df['Product_Category_3'].fillna(df['Product_Category_3'].mode()[0])
```

```python
df.head()
```

```python
df['Stay_In_Current_City_Years'].unique()
```

```python
df['Stay_In_Current_City_Years']=df['Stay_In_Current_City_Years'].str.replace('+','')
```

```python
df.head()
```

```python
##convert object into integer
# Fill missing values in 'Stay_In_Current_City_Years' with the mode before converting to int
df['Stay_In_Current_City_Years'] = df['Stay_In_Current_City_Years'].fillna(df['Stay_In_Current_City_Years'].mode()[0])
df['Stay_In_Current_City_Years']=df['Stay_In_Current_City_Years'].astype(int)
df.info()
```

```python
df['B']=df['B'].astype(int)
df['C']=df['C'].astype(int)
```
